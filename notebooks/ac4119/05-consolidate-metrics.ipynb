##  K-Nearest Neighbours

```python
drh_results = pd.read_csv('drh_results.csv')
drh_results.columns = 'n_neighbors', 'AUC', 'Average Precision'
```

```python
n_colors = 5
greens = sns.cubehelix_palette(n_colors, start=2.5, rot=-.3)
#sns.palplot(greens)
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("poster")
plt.figure(figsize=(20, 12))

sns.lineplot(x='n_neighbors', y='AUC', color = greens[2], data = drh_results  )\
.set_title('K-Nearest Neighbors')
```

```python
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))

sns.lineplot(x='n_neighbors', y='Average Precision', color = greens[2], data = drh_results  )\
.set_title('K-Nearest Neighbors')
```

```python
best_ap_score = max(drh_results['Average Precision'])
best_auc_score = max(drh_results['AUC'])
# ece_results 

drh_results[drh_results['Average Precision'] == best_ap_score]
```

```python
# SGD
ac_results = pd.read_csv('ac_results.csv')
ac_results.columns = ['Alpha', 'Loss', 'penalty', 'AUC', 'Average Precision']
n_colors = len(ac_results['Loss'].unique())
purples = sns.cubehelix_palette(n_colors, start=0.4, rot=-.1)
sns.palplot(purples)
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("poster")
plt.figure(figsize=(20, 12))
f, ax = plt.subplots(figsize=(20, 12))
ax.set(xscale="log")
sns.lineplot(x='Alpha', y='AUC', hue ='Loss',  palette = purples, data = ac_results, ci=None )\
.set_title('SGD Classifier')
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("poster")
plt.figure(figsize=(20, 12))
f, ax = plt.subplots(figsize=(20, 12))
ax.set(xscale="log")
sns.lineplot(x='Alpha', y='Average Precision', hue ='Loss',  palette = purples, data = ac_results, ci=None )\
.set_title('SGD Classifier')
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))
f, ax = plt.subplots(figsize=(5, 3))
ax.set(xscale="log")
sns.lineplot(x='Alpha', y='AUC', hue ='Loss',  palette = purples, data = ac_results, ci=None )\
.set_title('SGD Classifier')
```
