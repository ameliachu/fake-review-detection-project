##  K-Nearest Neighbours

```python
drh_results = pd.read_csv('drh_results.csv')
drh_results.columns = 'n_neighbors', 'AUC', 'Average Precision'
```

```python
n_colors = 5
greens = sns.cubehelix_palette(n_colors, start=2.5, rot=-.3)
#sns.palplot(greens)
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))

sns.lineplot(x='n_neighbors', y='AUC', color = greens[2], data = drh_results  )\
.set_title('K-Nearest Neighbors')
```

```python
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))

sns.lineplot(x='n_neighbors', y='Average Precision', color = greens[2], data = drh_results  )\
.set_title('K-Nearest Neighbors')
```

### Naive Bayes

```python
# Naive Bayes
ece_results = pd.read_csv('ece_results.csv')
ece_results.columns = ['alpha', 'fit_prior', 'Average Precision', 'AUC']
blues = sns.cubehelix_palette(5, start=0, rot=-.3)
blues = [blues[1], blues[3]]
sns.palplot(blues)
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))

sns.lineplot(x='alpha', y='AUC', hue ='fit_prior',  palette = blues, data = ece_results  )\
.set_title('Naive Bayes')
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))

sns.lineplot(x='alpha', y='Average Precision', hue ='fit_prior',  palette = blues, data = ece_results  )\
.set_title('Naive Bayes')
```

```python
# Random Forest
reds = sns.cubehelix_palette(4, start=1, rot=-.1)
gh_maxd_results = pd.read_csv('gh_maxd_results.csv')
gh_nest_results = pd.read_csv('gh_nest_results.csv')
gh_nsplit_results = pd.read_csv('gh_nsplit_results.csv')
gh_maxd_results.columns = 'Max Depth' , 'AUC', 'Average Precision'
gh_nest_results.columns = 'N Estimators', 'AUC', 'Average Precision'
gh_nsplit_results.columns = 'N Splits', 'AUC', 'Average Precision'
```

```python
reds = sns.cubehelix_palette(5, start=1, rot=-.1)
sns.palplot(reds)
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))

sns.lineplot(x='Max Depth', y='AUC', color = reds[2], data = gh_maxd_results)\
.set_title('Random Forest Classifier')
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))

sns.lineplot(x='Max Depth', y='Average Precision', color = reds[2], data =  gh_maxd_results )\
.set_title('Random Forest Classifier')
```

```python
%matplotlib inline
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(5, 3))

sns.lineplot(x='Max Depth', y='AUC', color = reds[2], data = gh_nest_results)\
.set_title('Random Forest Classifier')
```
